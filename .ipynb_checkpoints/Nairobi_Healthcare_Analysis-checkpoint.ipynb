# Nairobi Healthcare Accessibility Analysis
## Cell 2: Introduction and Assumptions
- **Objective**: Analyze health facilities and demographic data to identify gaps in healthcare access and propose investment priorities for Nairobi County, aligning with SDG 3 (Healthy Lives and Well-being).
- **Data Sources**:
  - Health Facilities Data: From [open.africa](https://open.africa/dataset/health-facilities-in-kenya).
  - 2019 Kenya Population and Housing Census: From [open.africa](https://open.africa/dataset/2019-kenya-population-and-housing-census).
- **Assumptions**:
  - Health facilities data, though from various years, is representative for 2025 analysis.
  - 2019 census data reflects current demographic trends.
  - Euclidean distances approximate travel time due to lack of road network data.
  - Analysis focuses on primary healthcare facilities (e.g., dispensaries, clinics).
- **Action**: This cell sets the context. No code execution is required. Proceed to the next cell after reviewing.

```python
def load_health_facilities():
    """
    Load and clean health facilities dataset from Excel file.
    Returns a pandas DataFrame with Nairobi primary healthcare facilities.
    """
    # Load Excel from URL
    url = "https://open.africa/dataset/3e95b5cb-39f5-44d3-94b6-f2d5285b0478/resource/0257f153-7228-49ef-b330-8e8ed3c7c7e8/download/ehealth-kenya-facilities-download-21102015.xls"
    try:
        df = pd.read_excel(url)
    except Exception as e:
        print(f"Error loading Excel file: {e}. Please check the URL or load the file locally.")
        return None
    
    # Clean column names (strip spaces and special characters)
    df.columns = df.columns.str.strip().str.replace(r'[*]', '', regex=True)
    
    # Print column names to verify
    print("Dataset Column Names:", df.columns.tolist())
    
    # Check if necessary columns exist
    if 'County' not in df.columns or 'Type' not in df.columns:
        print("Error: 'County' or 'Type' columns are missing in the dataset.")
        return None
    
    # Filter for Nairobi County
    df_nairobi = df[df['County'] == 'Nairobi'].copy()
    
    # Focus on primary healthcare facilities
    primary_types = ['Dispensary', 'Clinic', 'Health Centre']
    df_primary = df_nairobi[df_nairobi['Type'].isin(primary_types)]
    
    return df_primary

# Run the function and display summary
try:
    df_facilities = load_health_facilities()
    if df_facilities is not None:
        print(f"Number of primary healthcare facilities in Nairobi: {len(df_facilities)}")
        print(df_facilities[['Facility Name', 'District', 'Type']].head())
except KeyError as e:
    print(f"KeyError: {e}. Please check the column names above and adjust the code accordingly.")
```

# Nairobi Healthcare Accessibility Analysis
## Cell 3: Load and Preprocess Population Data
- **Objective**: Load 2019 Kenya Population and Housing Census data to analyze population, sex distribution, and household size by sub-county in Nairobi.
- **Data Sources**:
  - Population Sex Distribution by Sub-county: [CSV](https://open.africa/dataset/9b94fe50-9d75-4b92-be00-6354c6e6cc88/resource/ddddbe91-b5fa-4449-a50d-846401e701c4/download/distribution-of-population-by-sex-and-sub-county-2019-census-volume-ii.csv).
  - Average Household Size by Sub-county: [CSV](https://open.africa/dataset/9b94fe50-9d75-4b92-be00-6354c6e6cc88/resource/27107613-0f4b-481b-88ea-8c83165130d2/download/kenya-average-household-size-by-sub-county.csv).
- **Tasks**:
  - Load CSV files, adjusting for metadata rows (skiprows=2 for sex distribution, skiprows=6 for household size).
  - Filter for Nairobi sub-counties using a predefined list.
  - Merge datasets on sub-county to combine population, sex distribution, and household size.
  - Note: Land Area and Population Density missing; will be addressed in the next cell.
- **Action**: Run the code cell to load and process the data. Verify the output shows raw data, column names, and a table with sub-county, population, household size, and sex distribution for Nairobi. If URLs fail, download the files and load locally.

```python
def load_population_data():
    """
    Load and process 2019 Kenya Population and Housing Census data for Nairobi sub-counties.
    Returns a DataFrame with population, sex distribution, household size, land area, and population density.
    """
    # Load CSV files
    sex_url = "https://open.africa/dataset/9b94fe50-9d75-4b92-be00-6354c6e6cc88/resource/ddddbe91-b5fa-4449-a50d-846401e701c4/download/distribution-of-population-by-sex-and-sub-county-2019-census-volume-ii.csv"
    hh_url = "https://open.africa/dataset/9b94fe50-9d75-4b92-be00-6354c6e6cc88/resource/27107613-0f4b-481b-88ea-8c83165130d2/download/kenya-average-household-size-by-sub-county.csv"
    area_url = "https://open.africa/dataset/9b94fe50-9d75-4b92-be00-6354c6e6cc88/resource/4f04a378-0887-498c-b559-7ba000df35a0/download/distribution-of-population-by-land-area-density-and-sub-county-2019-census-volume-ii.csv"
    
    try:
        df_sex = pd.read_csv(sex_url, skiprows=2)
        df_hh = pd.read_csv(hh_url, skiprows=6)
        df_area = pd.read_csv(area_url)
    except Exception as e:
        print(f"Error loading CSV files: {e}. Please download the files from the URLs and load them locally.")
        return None
    
    # Clean column names
    df_sex.columns = df_sex.columns.str.strip()
    df_hh.columns = df_hh.columns.str.strip()
    df_area.columns = df_area.columns.str.strip()
    
    # Clean sub-county names
    df_sex['Sub County'] = df_sex['Sub County'].str.strip().str.replace(r'[*]', '', regex=True).str.title()
    df_hh['name'] = df_hh['name'].str.strip().str.replace(r'[*]', '', regex=True).str.title()
    df_area['Sub County'] = df_area['Sub County'].str.strip().str.replace(r'[*]', '', regex=True).str.title()

    nairobi_sub_counties = [
        'Dagoretti North', 'Dagoretti South', 'Embakasi East', 'Embakasi North', 
        'Embakasi South', 'Embakasi West', 'Kamukunji', 'Kasarani', 'Kibra', 
        'Langata', 'Makadara', 'Mathare', 'Roysambu', 'Ruaraka', 'Starehe', 
        'Westlands', 'Embakasi Central'
    ]

    df_sex_nairobi = df_sex[df_sex['Sub County'].isin(nairobi_sub_counties)].copy()
    df_hh_nairobi = df_hh[df_hh['name'].isin(nairobi_sub_counties)].copy()
    df_area_nairobi = df_area[df_area['Sub County'].isin(nairobi_sub_counties)].copy()

    # Rename columns
    df_sex_nairobi = df_sex_nairobi.rename(columns={
        'Sub County': 'Sub_County',
        'Total': 'Population',
        'Male': 'Male_Population',
        'Female': 'Female_Population'
    })
    df_hh_nairobi = df_hh_nairobi.rename(columns={
        'name': 'Sub_County',
        'avg_size': 'Household_Size'
    })
    df_area_nairobi = df_area_nairobi.rename(columns={
        'Sub County': 'Sub_County',
        'Sq. Km': 'Area_km2'
    })

    # Clean numeric columns (remove commas and convert to float)
    for col in ['Population', 'Male_Population', 'Female_Population']:
        df_sex_nairobi[col] = df_sex_nairobi[col].replace(',', '', regex=True).astype(float)
    df_area_nairobi['Area_km2'] = df_area_nairobi['Area_km2'].replace(',', '', regex=True).astype(float)

    # Merge dataframes
    df_merged = df_sex_nairobi[['Sub_County', 'Population', 'Male_Population', 'Female_Population']].merge(
        df_hh_nairobi[['Sub_County', 'Household_Size']], on='Sub_County', how='left'
    ).merge(
        df_area_nairobi[['Sub_County', 'Area_km2']], on='Sub_County', how='left'
    )

    # Calculate Population Density
    df_merged['Population_Density'] = df_merged['Population'] / df_merged['Area_km2']

    return df_merged

# Run and display
df_pop = load_population_data()
if df_pop is not None:
    print("\nPopulation Data for Nairobi Sub-Counties:")
    print(df_pop[['Sub_County', 'Population', 'Household_Size', 'Male_Population', 'Female_Population', 'Area_km2', 'Population_Density']].head())
```

# Nairobi Healthcare Accessibility Analysis
## Cell 4: Analyze Healthcare Accessibility and Visualize Results
- **Objective**: Analyze healthcare accessibility in Nairobi sub-counties using population, household size, land area, population density, and health facility data, and visualize with Matplotlib/Seaborn.
- **Data**:
  - Population Data: Merged DataFrame from Cell 4 with sub-county, population, household size, sex distribution, land area, and population density.
  - Health Facilities: 287 primary healthcare facilities in Nairobi (from Cell 3, assumed county-level).
- **Tasks**:
  - Impute missing `Household_Size` values (e.g., Langata) using the median of Nairobi sub-counties.
  - Distribute 287 health facilities across sub-counties using a weighted combination of population (70%) and area (30%).
  - Calculate accessibility metrics: facilities per 100,000 people and per km².
  - Visualize metrics using Seaborn barplots to identify disparities.
- **Action**: Run the code cell to impute missing data, calculate metrics, and display barplots. Verify the output shows a table with accessibility metrics and two barplots (facilities per 100,000 people and per km²). Ensure Matplotlib and Seaborn are installed (`pip install matplotlib seaborn`).

```python
def analyze_healthcare_access(df_pop, total_facilities=287, pop_weight=0.7, area_weight=0.3):
    """
    Analyze healthcare accessibility in Nairobi sub-counties and visualize with Matplotlib/Seaborn.
    Parameters:
        df_pop (DataFrame): Merged population data from Cell 4.
        total_facilities (int): Total number of primary healthcare facilities (287).
        pop_weight (float): Weight for population-based allocation (default 0.7).
        area_weight (float): Weight for area-based allocation (default 0.3).
    Returns:
        DataFrame with accessibility metrics and displays barplots.
    """
    # Impute missing Household_Size with median
    df_pop['Household_Size'] = df_pop['Household_Size'].fillna(df_pop['Household_Size'].median())
    
    # Distribute facilities using weighted combination of population and area
    total_population = df_pop['Population'].sum()
    total_area = df_pop['Area_km2'].sum()
    df_pop['Facilities_By_Population'] = (df_pop['Population'] / total_population * total_facilities * pop_weight).round(2)
    df_pop['Facilities_By_Area'] = (df_pop['Area_km2'] / total_area * total_facilities * area_weight).round(2)
    df_pop['Facilities_Total'] = df_pop['Facilities_By_Population'] + df_pop['Facilities_By_Area']
    
    # Calculate accessibility metrics
    df_pop['Facilities_Per_100k'] = (df_pop['Facilities_Total'] / df_pop['Population'] * 100000).round(2)
    df_pop['Facilities_Per_km2'] = (df_pop['Facilities_Total'] / df_pop['Area_km2']).round(2)
    
    # Display metrics
    print("Healthcare Accessibility Metrics for Nairobi Sub-Counties:")
    print(df_pop[['Sub_County', 'Population', 'Area_km2', 'Population_Density', 'Household_Size', 'Facilities_Total', 'Facilities_Per_100k', 'Facilities_Per_km2']])
    
    # Set Seaborn style
    sns.set_style("whitegrid")
    
    # Plot 1: Facilities per 100,000 people
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Sub_County', y='Facilities_Per_100k', data=df_pop, color='#1f77b4')
    plt.title('Healthcare Facilities per 100,000 People by Sub-County')
    plt.xlabel('Sub-County')
    plt.ylabel('Facilities per 100,000 People')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    
    # Plot 2: Facilities per km²
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Sub_County', y='Facilities_Per_km2', data=df_pop, color='#ff7f0e')
    plt.title('Healthcare Facilities per km² by Sub-County')
    plt.xlabel('Sub-County')
    plt.ylabel('Facilities per km²')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    
    return df_pop

# Assuming df_pop is available from Cell 4
try:
    df_access = analyze_healthcare_access(df_pop)
except NameError:
    print("Error: df_pop not found. Please run Cell 4 first to generate the population DataFrame.")
```

#airobi Healthcare Accessibility Analysis
## Cell 5: Final Analysis and Prioritization
- **Objective**: Finalize the healthcare accessibility analysis by computing additional metrics and identifying sub-counties needing priority intervention, visualized with Matplotlib/Seaborn.
- **Data**:
  - Accessibility Metrics: From Cell 5, including population, household size, area, density, and facility counts per sub-county.
  - Total Facilities: 287 primary healthcare facilities in Nairobi.
- **Tasks**:
  - Calculate population per facility to assess access strain.
  - Identify sub-counties with access gaps (e.g., facilities per 100,000 below WHO’s ~10 per 100,000 threshold).
  - Visualize priority areas using a sorted barplot.
  - Provide final recommendations based on findings.
- **Action**: Run the code cell to compute metrics, display a sorted barplot, and finalize the analysis. Verify the output shows a table with additional metrics and a barplot highlighting priority sub-counties. Ensure Matplotlib and Seaborn are installed (`pip install matplotlib seaborn`).

```python
def finalize_healthcare_analysis(df_access):
    """
    Finalize healthcare accessibility analysis by computing additional metrics and identifying priority sub-counties.
    Parameters:
        df_access (DataFrame): Accessibility metrics from Cell 5.
    Returns:
        DataFrame with additional metrics and displays a sorted barplot.
    """
    # Calculate population per facility
    df_access['Population_Per_Facility'] = (df_access['Population'] / df_access['Facilities_Total']).round(2)
    
    # Identify access gaps (WHO recommends ~1 facility per 10,000 people, or 10 per 100,000)
    threshold = 10  # Facilities per 100,000
    df_access['Access_Gap'] = df_access['Facilities_Per_100k'].apply(lambda x: 'Underserved' if x < threshold else 'Adequate')
    
    # Display final metrics
    print("Final Healthcare Accessibility Metrics for Nairobi Sub-Counties:")
    print(df_access[['Sub_County', 'Population', 'Area_km2', 'Population_Density', 'Household_Size', 'Facilities_Total', 'Facilities_Per_100k', 'Facilities_Per_km2', 'Population_Per_Facility', 'Access_Gap']])
    
    # Set Seaborn style
    sns.set_style("whitegrid")
    
    # Plot: Sorted Facilities per 100,000 people with access gap annotation
    plt.figure(figsize=(12, 6))
    df_sorted = df_access.sort_values('Facilities_Per_100k', ascending=True)
    sns.barplot(x='Sub_County', y='Facilities_Per_100k', hue='Access_Gap', dodge=False, data=df_sorted, palette={'Underserved': '#d62728', 'Adequate': '#2ca02c'})
    plt.axhline(y=threshold, color='black', linestyle='--', label=f'WHO Threshold ({threshold} per 100k)')
    plt.title('Healthcare Facilities per 100,000 People by Sub-County (Sorted)')
    plt.xlabel('Sub-County')
    plt.ylabel('Facilities per 100,000 People')
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Access Status')
    plt.tight_layout()
    plt.show()
    
    return df_access

# Assuming df_access is available from Cell 5
try:
    df_final = finalize_healthcare_analysis(df_access)
except NameError:
    print("Error: df_access not found. Please run Cell 5 first to generate the accessibility DataFrame.")
```
